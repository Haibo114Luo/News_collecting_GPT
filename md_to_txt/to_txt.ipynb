```python
import os

# 设定目标文件夹路径 (根据您的截图，可能是 './News/AIDC' 或其他路径)
source_folder = [r'E:/风电数据库/News/AIDC', r'E:/风电数据库/News/Wind']
output_file = 'merged_context.txt'
```

```python
def merge_folders_from_list(folder_list, output_path):
    # 打开目标文件准备写入
    with open(output_path, 'w', encoding='utf-8') as outfile:
        
        # 第一层循环：遍历文件夹列表中的每一个文件夹路径
        for folder_path in folder_list:
            
            # 检查文件夹是否存在
            if os.path.exists(folder_path):
                folder_name = os.path.basename(folder_path)
                print(f"正在处理文件夹: {folder_name} ...")
                
                # 获取该文件夹下的文件列表
                # os.listdir() 在这里接收的是 folder_path (字符串)，所以不会报错了
                files = sorted(os.listdir(folder_path))
                
                # 第二层循环：遍历文件
                for filename in files:
                    if filename.endswith(".md") or filename.endswith(".txt"):
                        filepath = os.path.join(folder_path, filename)
                        
                        # 写入分隔符，标记来源文件夹和文件名
                        header = f"\n\n{'='*40}\nSOURCE: {folder_name} | FILE: {filename}\n{'='*40}\n\n"
                        outfile.write(header)
                        
                        try:
                            with open(filepath, 'r', encoding='utf-8') as infile:
                                outfile.write(infile.read())
                        except Exception as e:
                            print(f"读取错误 {filename}: {e}")
            else:
                print(f"警告: 找不到文件夹路径 -> {folder_path}")

    print(f"\n全部合并完成！输出文件: {output_path}")
```

```python
# 执行函数
merge_folders_from_list(source_folder, output_file)
```
